```python
all_songs = {}
ATTRIBUTES = ['danceability', 'energy', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence']
COLUMNS = ['country', 'name', 'artists', 'genre'] + ATTRIBUTES
index = 1
for country in TOP_PLAYLIST_DICT:
    songs = TOP_PLAYLIST_DICT[country]['tracks']['items']
    for song in songs:
        artist_id = song['track']['artists'][0]['id']
        artist_data = spotify.artist(artist_id)
        genre = artist_data['genres'][0] if artist_data['genres'] else None
        data = [country, song['track']['name'], song['track']['artists'][0]['name'], genre] 
        features = spotify.audio_features(tracks=song['track']['id'])
        for attribute in ATTRIBUTES:
            data.append(features[0][attribute])
        all_songs[index] = data
        index += 1

    
```

```python
import pandas as pd
```

```python
df = pd.DataFrame.from_dict(all_songs, orient='index',
                       columns=COLUMNS)
df.to_csv("../data/top50bycountry.csv")
```
